```python
## liking posts

# auto-scrolling not required as 30 posts show up initially 

base_url = driver.current_url

for i in range(30):
    
    driver.implicitly_wait(3)
    
    ## explicitly wait for a post to be visible
    post_btn = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[contains(@class, '_9AhH0')]")))[i]
    post_btn.click()
    
    ## explicitly wait for like button to be available
    like_button = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//span[@class='fr66n']/button[contains(@class, 'wpO6b')]")))[0]
    
    unliked_post_color_code = "#262626"
    
    like_status = driver.find_element_by_xpath("//button[contains(@class, 'wpO6b')][1]/div/span")
    
    if unliked_post_color_code in like_status.get_attribute("innerHTML"):
        driver.execute_script("arguments[0].click()", like_button)
        
    else:
        print("post", (i+1), "is already liked")
    
    webdriver.ActionChains(driver).send_keys(Keys.ESCAPE).perform()
    
    
```

```python
## un-liking posts

# auto-scrolling not required as 30 posts show up initially 

base_url = driver.current_url

for i in range(30):
    
    driver.implicitly_wait(3)
    
    ## explicitly wait for a post to be visible
    post_btn = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[contains(@class, '_9AhH0')]")))[i]
    post_btn.click()
    
    ## explicitly wait for like button to be available
    like_button = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//span[@class='fr66n']/button[contains(@class, 'wpO6b')]")))[0]
    
    unliked_post_color_code = "#262626"
    
    like_status = driver.find_element_by_xpath("//button[contains(@class, 'wpO6b')][1]/div/span")
    
    if unliked_post_color_code not in like_status.get_attribute("innerHTML"):
        driver.execute_script("arguments[0].click()", like_button)
        
    else:
        print("post", (i+1), "is already un-liked")
    
    webdriver.ActionChains(driver).send_keys(Keys.ESCAPE).perform()
    
    
```

```python
# get followers
followers_element.click()
time.sleep(1)
followers_scroller = driver.find_element_by_xpath("//ul[contains(@class, 'jSC57')]")

while True:
    
    if(len(driver.find_elements_by_xpath("//a[contains(@class, 'FPmhX notranslate')]")) == followers_num):
        break
        
    driver.execute_script("arguments[0].scrollIntoView(0,100);", followers_scroller)
    time.sleep(0.5)
    
my_followers = [name.get_attribute("title") for name in 
                driver.find_elements_by_xpath("//a[contains(@class, 'FPmhX notranslate')]")]
```

```python
# get following
driver.back()
time.sleep(1)

following_element = driver.find_element_by_partial_link_text("following")
following_num = int(driver.find_element_by_xpath("//li[3]/a/span").get_attribute("innerHTML").strip())
me_following = []

following_element.click()
time.sleep(1)
following_scroller = driver.find_element_by_xpath("//ul[contains(@class, 'jSC57')]")

while True:
    
    if(len(driver.find_elements_by_xpath("//a[contains(@class, 'FPmhX notranslate')]")) == following_num):
        break
        
    driver.execute_script("arguments[0].scrollIntoView(0,100);", following_scroller)
    time.sleep(0.5)
    
me_following = [name.get_attribute("title") for name in 
                driver.find_elements_by_xpath("//a[contains(@class, 'FPmhX notranslate')]")]

my_followers = set(my_followers)
me_following = set(me_following)
food_talk_india = set(food_talk_india)
```
